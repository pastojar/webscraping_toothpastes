### Connect and save the webpage data

```python
url_notino = "https://www.notino.co.uk/toothpaste/"
```

```python
# Set up the Chrome WebDriver
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service)  

# Open the webpage
driver.get(url_notino)
```

Make the webpage show you all the tothpaste items

```python
# Get the page source
page_source = driver.page_source

# Capture the current timestamp
scraping_time = datetime.now()

# Parse the page source with BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')
```

```python
# Close the WebDriver
driver.quit()
```

----

### Extract the relevant data

```python
# Create an empty data frame
col_names = ["url", "image_url", "brand", "product", "add_info", "price_value", "price_currency", "discount_value", "discount_code"]
toothpaste_data = pd.DataFrame( columns=col_names)
```

```python
# Populate the data frame with appropriate values
for item in soup.find_all("a", class_ = "sc-jdHILj OFtqG"):
    new_item = {key: None for key in col_names}
    
    new_item[ col_names[0] ] = "https://www.notino.co.uk" + str( item["href"] )
    new_item[ col_names[1] ] = str( item.find("img")["src"] )    
    new_item[ col_names[2] ] = str( item.find("h2").string )
    new_item[ col_names[3] ] = str( item.find("h3").string )    
    new_item[ col_names[4] ] = str( item.find("p").string )    
    if item.find("span", class_="sc-hhyKGa sc-gYrqIg iwwcvf dOVzXY" ) :
        new_item[ col_names[5] ] = ( item.find("span", class_="sc-hhyKGa sc-gYrqIg iwwcvf dOVzXY" ).string )
    if item.find("span", class_="sc-hhyKGa sc-cCzLxZ jOWcPO hsLJgo" ) :
        new_item[ col_names[6] ] = str( item.find("span", class_="sc-hhyKGa sc-cCzLxZ jOWcPO hsLJgo" ).string )    
    if item.find("span", class_="styled__DiscountValue-sc-1b3ggfp-1 jWXmOz" ) :
        new_item[ col_names[7] ] = ( item.find("span", class_="styled__DiscountValue-sc-1b3ggfp-1 jWXmOz" ).string )    
    if item.find("span", class_="styled__StyledDiscountCode-sc-1i2ozu3-1 gfxrfw" ) :
        new_item[ col_names[8] ] = str( item.find("span", class_="styled__StyledDiscountCode-sc-1i2ozu3-1 gfxrfw" ).string )  
        
    toothpaste_data.loc[len(toothpaste_data)] = new_item
```

```python
# Define appropriate data types
toothpaste_data[["price_value"]] = toothpaste_data[["price_value"]].apply(pd.to_numeric)
```
